This notebook illustrates how to run a Neurokernel-based model comprising partial models of the fruit fly's two antennal lobes. First, we generate an input and execute the model:

```python
%cd ~/neurokernel/examples/olfaction_integration/data
%run gen_olf_input.py
%cd ..
%run olfaction_integration_demo.py
```

Next, we display the spikes produced by all of the neurons in the model. In the diagrams below, 
non-zero input and output spikes are white:

```python
import matplotlib.ticker as ticker
fmt = lambda x, pos: '%2.2f' % (float(x)/1e4)
with h5py.File('./data/olfactory_input.h5', 'r') as fi, \
    h5py.File('olfactory_left_output_spike.h5','r') as fo_left, \
    h5py.File('olfactory_right_output_spike.h5', 'r') as fo_right:
    data_i = fi['real'].value;
    data_o_left = fo_left['array'].value;
    data_o_right = fo_right['array'].value;
rcParams['figure.dpi']=120; rcParams['figure.figsize']=(12,9)
subplot(311); set_cmap('gray');
gca().xaxis.set_major_formatter(ticker.FuncFormatter(fmt))
imshow(data_i.T); axis('tight'); title('Input'); ylabel('Neuron #'); 
subplot(312); set_cmap('gray');
gca().xaxis.set_major_formatter(ticker.FuncFormatter(fmt))
imshow(data_o_left.T); axis('tight'); title('Output (left)'); ylabel('Neuron #');
subplot(313); set_cmap('gray');
gca().xaxis.set_major_formatter(ticker.FuncFormatter(fmt))
imshow(data_o_right.T); axis('tight'); title('Output (right)'); ylabel('Neuron #');
xlabel('time (s)')
subplots_adjust();
```
