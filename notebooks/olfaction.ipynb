Display the spikes produced by all of the neurons in the model. In the diagrams below, 
non-zero input and output spikes are white:

```python
import matplotlib.ticker as ticker
fmt = lambda x, pos: '%2.2f' % (float(x)/1e4)
with h5py.File('./data/olfactory_input.h5', 'r') as fi, \
    h5py.File('olfactory_output_spike.h5','r') as fo:
    data_i = fi['real'].value; print data_i.shape
    data_o = fo['array'].value; print data_o.shape
rcParams['figure.dpi']=120; rcParams['figure.figsize']=(12,6)
subplot(211); set_cmap('gray');
gca().xaxis.set_major_formatter(ticker.FuncFormatter(fmt))
imshow(data_i.T); axis('tight'); title('Input'); ylabel('Neuron #'); 
subplot(212); set_cmap('gray');
gca().xaxis.set_major_formatter(ticker.FuncFormatter(fmt))
imshow(data_o.T); axis('tight'); title('Output'); ylabel('Neuron #');
xlabel('time (s)')
subplots_adjust();
```
